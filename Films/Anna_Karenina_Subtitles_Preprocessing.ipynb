# Достаем прямую и косвенную речь из оригинального текста

## - Это из целикового текста, на выходе - один единый документ

```python
with open('/Users/tonleon/Downloads/anna-karenina.txt', encoding='utf8') as file:
    text_of_AK = file.readlines()
    file.close()
```

```python
def get_direct_and_reported_speech_from_text(list_of_text_lines):
    direct_reported_speech = []
    if_start_undirect = False
    for line in list_of_text_lines:

        line = re.sub('\[.*?\]', '', line)

        if "«" in line:

            if '»' not in line:
                if_start_undirect = True
                lindex = line.find("«")
                current_undirect_string = line[lindex:]
            else:
                reported = re.findall('\«.+?\»', line)
                direct_reported_speech.append(reported)
    #             lindex, rindex = line.find("«"), line.rfind("»")
    #             direct_reported_speech.append(line[lindex: rindex + 1])
        elif if_start_undirect:
            if '»' in line:
                rindex = line.rfind("»")
                current_undirect_string += line[:rindex + 1]
                direct_reported_speech.append(current_undirect_string)
                if_start_undirect = False
            else:
                current_undirect_string += line
        elif len(line) > 0 and line[0] == '–':
            line = re.sub('–.[а-яё]+.+?\s–|–.[а-яё]+.+', '', line)
            direct_reported_speech.append(line)
            
    return direct_reported_speech
```

```python
direct_reported_speech = get_direct_and_reported_speech_from_text(text_of_AK)
direct_reported_speech
```

```python
final_text = []
for line in direct_reported_speech:
    if isinstance(line, str):
        final_text.append(line)
    if isinstance(line, list):
        for element in line:
            final_text.append(element)

final_text
with open("Reported_and_Directed_speech_in_Anna_Karenina.txt", "w") as file:
    for line in final_text:
        file.write(str(line+'\n'))
```

## - Это из текста, поделенного на 8 частей, на выходе - 8 документов с речью прямой и косвенной для каждой части

```python
with open('/Users/tonleon/Downloads/anna-karenina.txt', encoding='utf8') as file:
    text_of_AK_splited = file.read()
    file.close()
    text_of_AK_splited = text_of_AK_splited.split('Часть')
```

```python
part_1 = text_of_AK_splited[1].split('\n')
part_2 = text_of_AK_splited[2].split('\n')
part_3 = text_of_AK_splited[3].split('\n')
part_4 = text_of_AK_splited[4].split('\n')
part_5 = text_of_AK_splited[5].split('\n')
part_6 = text_of_AK_splited[6].split('\n')
part_7 = text_of_AK_splited[7].split('\n')
part_8 = text_of_AK_splited[8].split('\n')
```

```python
for i, text in enumerate([part_1, part_2, part_3, part_4, part_5, part_6, part_7, part_8]):
    direct_reported_speech_parts = get_direct_and_reported_speech_from_text(text)
    final_text = []
    for line in direct_reported_speech_parts:
        if isinstance(line, str):
            final_text.append(line)
        if isinstance(line, list):
            for element in line:
                final_text.append(element)
    with open("Reported_and_Directed_speech_in_Anna_Karenina_Part_{}.txt".format(i+1), "w") as file:
        for line in final_text:
            file.write(str(line+'\n'))
```

```python
def get_clean_subtitles(original_subtitles):
    sub = []
    punct = [".","!","?","..."]
    for line in original_subtitles:
        line = re.sub('\d+.\d+.+\d+.\d+.+|\d.+', '', line)
        if line != '\n':
            sub.append(line)
    whole_text = ''.join(sub)
    whole_text = re.sub('\n', ' ', whole_text)
    whole_text = re.sub('[1-9]', '', whole_text)
    return whole_text
```

```python
def save_subtitles(subtitles, year):
    with open("Subtitles_{}.txt".format(year), "w") as file:
        file.write(subtitles)
```

```python
sub_1967 = get_clean_subtitles(subtitles_1967)
sub_1997 = get_clean_subtitles(subtitles_1997)
sub_1997 = re.sub('\.{3}\s+\.{3}|\,\.{2}\s+\.{3}', ' ', sub_1997)
sub_2012 = get_clean_subtitles(subtitles_2012)
sub_2012
```

```python
save_subtitles(sub_1997, 1997)
save_subtitles(sub_1967, 1967)
save_subtitles(sub_2012, 2012)
```
