```python
'''    '''

def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
'''      xml,
 ,   , ,  
  (   , 
   xml )'''

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_faces(cascade, image, scaleFactor = 1.2):
    '''     '''
    image_copy = image.copy()

    '''    ,   detectMultiScale
            '''
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    '''    haar classifier''' 
    faces = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=5)
    
    print(" : " + format(len(faces)))

    for (x, y, w, h) in faces:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 5)

    return image_copy
```

```python
faces = detect_faces(face_cascade, image)
plt.imshow(convertToRGB(faces))
```
